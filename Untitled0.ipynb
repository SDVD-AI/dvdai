````python
import requests
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

def generate_ai_html(b):
    user_prompt = input_text.value.strip()
    if not user_prompt:
        print("אנא כתוב תיאור לאתר (למשל: אתר למכירת פיצה בצבע כחול)")
        return

    with output_area:
        clear_output()
        print(f"🤖 ה-AI חושב על עיצוב עבור: '{user_prompt}'...")

        # בניית השאלה לבינה המלאכותית
        system_msg = "You are an expert HTML/CSS developer. Return ONLY the code for a single, complete HTML file with embedded CSS."
        full_prompt = f"{system_msg} User request: Create a professional website for: {user_prompt}"

        # שליחת הבקשה ל-AI (דרך Pollinations API)
        url = f"https://text.pollinations.ai/{full_prompt.replace(' ', '%20')}"

        try:
            response = requests.get(url)
            html_code = response.text

            # ניקוי סימני Markdown אם ה-AI הוסיף אותם
            clean_code = html_code.replace("```html", "").replace("```", "").strip()

            # הצגת התוצאה
            display(HTML(f"<h3 style='color: #00ff00;'>✨ האתר שלך מוכן!</h3>"))

            # כפתור להצגת הקוד
            print("--- הנה הקוד שנוצר (העתק אותו לקובץ .html): ---")
            print(clean_code)

            # תצוגה חיה של האתר בתוך ה-Colab!
            display(HTML("<hr><h4>תצוגה מקדימה:</h4>"))
            display(HTML(f"<div style='border: 5px solid gray; padding: 10px;'>{clean_code}</div>"))

        except Exception as e:
            print(f"❌ שגיאה בחיבור ל-AI: {e}")

# עיצוב הממשק
input_text = widgets.Text(placeholder='תאר את האתר שאתה רוצה...', layout=widgets.Layout(width='70%'))
gen_btn = widgets.Button(description="צור אתר עם AI", button_style='primary')
output_area = widgets.Output()

gen_btn.on_click(generate_ai_html)

display(HTML("<h1 style='color: #3498db;'>🚀 AI Web Designer</h1>"))
display(widgets.HBox([input_text, gen_btn]))
display(output_area)
````
